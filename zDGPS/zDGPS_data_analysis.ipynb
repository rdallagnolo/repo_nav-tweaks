```python
df = pd.read_csv('V1G1report.2021_021',skiprows=4)
```

```python
sns.set_style('whitegrid')
fig, axes = plt.subplots(1,3,figsize=(15,5))

pitch=df['pitch']
roll=df['roll']
heave=df['heave']

sns.distplot(pitch,ax=axes[0],color='black',bins=20,kde=False)
axes[0].set_title('pitch',fontsize=20)
axes[0].set_ylabel('count',fontsize=20)
axes[0].set_xlabel('pitch',fontsize=20)

sns.histplot(roll,ax=axes[1],bins=20,kde=False,palette='autumn')
axes[1].set_title('roll',fontsize=20)
axes[1].set_ylabel('count',fontsize=20)
axes[1].set_xlabel('roll',fontsize=20)

sns.histplot(heave,ax=axes[2],bins=20,kde=False, palette='seismic')
axes[2].set_title('heave',fontsize=20)
axes[2].set_ylabel('count',fontsize=20)
axes[2].set_xlabel('heave',fontsize=20)

plt.tight_layout()
```

```python
sns.set_style('whitegrid')
fig1, axes1 = plt.subplots(2,2,figsize=(10,10))

sns.regplot(x='raw-lat',y='proc-lat',data=df,ax=axes1[0,0])
axes1[0,0].set_title('Lat transformation',fontsize=20)

sns.regplot(x='raw-lon',y='proc-lon',data=df,ax=axes1[0,1])
axes1[0,1].set_title('Long transformation',fontsize=20)

sns.regplot(x='raw-geo-ht',y='proc-geo-ht',data=df,ax=axes1[1,0])
axes1[1,0].set_title('Geoid height transformation',fontsize=20)

sns.regplot(x='raw-sph-ht',y='proc-sph-ht',data=df,ax=axes1[1,1])
axes1[1,1].set_title('Spheroid height transformation',fontsize=20)

plt.tight_layout()
```

```python
sns.set_style('whitegrid')
fig2, axes2 = plt.subplots(2,2,figsize=(10,10))

sph_proc=df['proc-sph-ht']
sph_raw=df['raw-sph-ht']
geo_proc=df['proc-geo-ht']
geo_raw=df['raw-geo-ht']


sns.distplot(sph_proc,ax=axes2[0,0])
axes2[0,0].set_title('Processed Spheroid Height',fontsize=20)

sns.distplot(sph_raw,ax=axes2[0,1])
axes2[0,1].set_title('Raw Spheroid Height',fontsize=20)

sns.distplot(geo_proc,ax=axes2[1,0])
axes2[1,0].set_title('Processed Geoid Height',fontsize=20)

sns.distplot(geo_raw,ax=axes2[1,1])
axes2[1,1].set_title('Raw Geoid Height',fontsize=20)

plt.tight_layout()
```
